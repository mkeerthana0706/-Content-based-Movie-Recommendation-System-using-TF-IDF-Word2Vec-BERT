# Data Preprocessing

```python
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
import re

# Load the CSV file
df = pd.read_csv("/content/sample_data/netflix.csv")

# Extract the 'description' column from the dataframe
description = df['description'].tolist()

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize WordNetLemmatizer
lemma = WordNetLemmatizer()

corpus = []

for des in description:
    des = re.sub('[^a-zA-Z]',' ', des)
    des = des.lower()
    des = des.split()
    des = [lemma.lemmatize(word) for word in des if word not in set(stopwords.words("english"))]
    des = " ".join(des)
    corpus.append(des)
```

```python
df["corpus"] = corpus
```

# Content-based recommendation system using TF-IDF vectorization

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Load the data
data = pd.read_csv('cleaned_netflixtitles_corpus.csv')

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# TF-IDF Vectorization
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(train_data['corpus'])

# Calculate cosine similarity
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to get movie recommendations based on user's interests and ratings
def get_recommendations(user_id, cosine_sim=cosine_sim, data=train_data):
    user_ratings = data[data['user_id'] == user_id]['rating']
    if user_ratings.empty:
        return []  # No rated movies for this user

    user_mean_rating = user_ratings.mean()
    idx = data[data['user_id'] == user_id].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = [(i, score * user_ratings[i]) for i, score in sim_scores if i in user_ratings.index]

    if not sim_scores:
        return []  # No similar movies found

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    return data['title'].iloc[movie_indices].tolist()

# Example recommendation for a user who rated movies
print("Recommendations for user 7:", get_recommendations(7))

# Evaluate the model
def evaluate_model(model, test_data):
    y_true = test_data['rating']
    y_pred = []
    for idx, row in test_data.iterrows():
        movies = train_data[train_data['user_id'] == row['user_id']]
        if not movies.empty:
            y_pred.append(movies['rating'].mean())
        else:
            y_pred.append(train_data['rating'].mean())

    y_pred = np.array(y_pred)
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    return mse, mae, rmse

# Evaluate the model
mse, mae, rmse = evaluate_model(cosine_sim, test_data)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```

In this model, the recommendations are based on the textual content of the movies' descriptions (represented by the corpus column) using TF-IDF vectorization. The model calculates the cosine similarity between movies' descriptions to find similar movies. Additionally, the model takes into account the user's ratings to personalize the recommendations. The rating column is used to calculate the weighted similarity scores, where movies with higher ratings contribute more to the recommendation score for a given user.

# Content-based recommendation system using Word2Vec

```python
from gensim.models import Word2Vec
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv('cleaned_netflixtitles_corpus.csv')

# Preprocess the text
corpus = data['corpus'].apply(lambda x: x.split())

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Reset index of train_data and test_data
train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)

# Train Word2Vec model
word2vec_model = Word2Vec(corpus, vector_size=100, window=5, min_count=1, sg=0)

# Function to calculate average word embeddings for a sentence
def avg_word_vectors(sentence, model, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    n_words = 0
    for word in sentence:
        if word in model.wv.index_to_key:
            n_words += 1
            feature_vector = np.add(feature_vector, model.wv[word])
    if n_words > 0:
        feature_vector = np.divide(feature_vector, n_words)
    return feature_vector

# Create movie embeddings
embedding_size = 100
movie_embeddings = [avg_word_vectors(sentence, word2vec_model, embedding_size) for sentence in corpus]

# Function to get movie recommendations based on user's interests and ratings
def get_recommendations(user_id, embeddings=movie_embeddings, data=train_data):
    user_ratings = data[data['user_id'] == user_id]['rating']
    if user_ratings.empty:
        return []  # No rated movies for this user

    user_mean_rating = user_ratings.mean()
    idx = data[data['user_id'] == user_id].index[0]
    sim_scores = cosine_similarity([embeddings[idx]], embeddings)[0]
    sim_scores = [(i, score * user_ratings[i]) for i, score in enumerate(sim_scores) if i in user_ratings.index]

    if not sim_scores:
        return []  # No similar movies found

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    return data['title'].iloc[movie_indices].tolist()

# Example recommendation for a user who rated movies
print("Recommendations for user 7:", get_recommendations(7))

# Evaluate the model
def evaluate_model(embeddings, test_data):
    y_true = test_data['rating']
    y_pred = []
    for idx, row in test_data.iterrows():
        movies = train_data[train_data['user_id'] == row['user_id']]
        if not movies.empty:
            movie_idx = movies.index[0]
            sim_scores = cosine_similarity([embeddings[movie_idx]], embeddings)[0]
            sim_scores = [(i, score * row['rating']) for i, score in enumerate(sim_scores) if i in movies.index]
            if sim_scores:
                y_pred.append(sum([score for _, score in sim_scores]) / len(sim_scores))
            else:
                y_pred.append(movies['rating'].mean())
        else:
            y_pred.append(train_data['rating'].mean())

    y_pred = np.array(y_pred)
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    return mse, mae, rmse

# Evaluate the model
mse, mae, rmse = evaluate_model(movie_embeddings, test_data)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```


In this Word2Vec-based recommendation model, the attributes considered for recommendation include the following:

User ID: The ID of the user for whom recommendations are being generated.
Movie Embeddings: The Word2Vec embeddings of movies, which capture the semantic meanings of movie descriptions.
Movie Ratings: The ratings given by users to movies, which are used to weigh the similarity scores between movies.
Movie Titles: The titles of movies, which are used for display purposes in the recommendations.
The model calculates the similarity between the user's rated movies and all other movies based on their embeddings. It then uses the ratings given by the user to weigh these similarities and recommend the most similar movies. The recommendation is based on the average rating-weighted similarity scores of the top similar movies.


# Content-based recommendation system using BERT technique

```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
import numpy as np

# Load pre-trained BERT model and tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# Load the data
data = pd.read_csv('cleaned_netflixtitles_corpus.csv')

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Reset index of train_data and test_data
train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)

# Function to get BERT embeddings for a list of descriptions
def get_bert_embeddings(descriptions, max_length=128):
    if isinstance(descriptions, str):  # Single example
        descriptions = [descriptions]
    inputs = tokenizer(descriptions, padding=True, truncation=True, max_length=max_length, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)
    if 'last_hidden_state' in outputs:
        last_hidden_state = outputs.last_hidden_state
        # Use the embeddings for [CLS] token (first token) as the sentence embedding
        embeddings = last_hidden_state[:, 0, :]
        return embeddings.numpy()
    else:
        raise ValueError("No 'last_hidden_state' found in model outputs.")

# Create movie embeddings
movie_embeddings = get_bert_embeddings(train_data['corpus'].tolist())

# Function to get movie recommendations based on user's interests and ratings
def get_recommendations(user_id, embeddings=movie_embeddings, data=train_data):
    user_ratings = data[data['user_id'] == user_id]['rating']
    if user_ratings.empty:
        return []  # No rated movies for this user

    user_mean_rating = user_ratings.mean()
    idx = data[data['user_id'] == user_id].index[0]
    sim_scores = cosine_similarity([embeddings[idx]], embeddings)[0]
    sim_scores = [(i, score * user_ratings[i]) for i, score in enumerate(sim_scores) if i in user_ratings.index]

    if not sim_scores:
        return []  # No similar movies found

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    return data['title'].iloc[movie_indices].tolist()

# Example recommendation for a user who rated movies
print("Recommendations for user 1:", get_recommendations(1))

# Evaluate the model
def evaluate_model(embeddings, test_data):
    y_true = test_data['rating']
    y_pred = []
    for idx, row in test_data.iterrows():
        movies = train_data[train_data['user_id'] == row['user_id']]
        if not movies.empty:
            movie_idx = movies.index[0]
            sim_scores = cosine_similarity([embeddings[movie_idx]], embeddings)[0]
            sim_scores = [(i, score * row['rating']) for i, score in enumerate(sim_scores) if i in movies.index]
            if sim_scores:
                y_pred.append(sum([score for _, score in sim_scores]) / len(sim_scores))
            else:
                y_pred.append(movies['rating'].mean())
        else:
            y_pred.append(train_data['rating'].mean())

    y_pred = np.array(y_pred)
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    return mse, mae, rmse

# Evaluate the model
mse, mae, rmse = evaluate_model(movie_embeddings, test_data)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```

This code first loads the pre-trained BERT model and the tokenizer for the purpose of generating embeddings for movie descriptions and also divides the data into the training and testing datasets for evaluating a basic content-based recommendation system using these embeddings. The results are generated by calculating cosine similarity of the BERT embeddings as well as the MSE, MAE and RMSE to validate the performance of the model.
